In this notebook,we implement the formulas for the exponential generating series 
$$
g_m(z, t) = \sum_{n, r} R_{m,n, r}(x) \frac{z^n}{n!} \frac{t^r}{r!} 
$$
for the rook polynomials on a tridimensional $m \times n \times r$ chessboard, for $m$ up to $3$.

There rook polynomials are themselves the ordinary generating series for the numbers of placements $S \in R(m, n, r| d)$ of $d$ non-attacking rooks:

$$
R_{m,n, r}(x)= 
\sum_d \# R(m,n,r|d) \cdot  x^d
$$

Note that $R(m,n,r|d)$ can be also interpreted as the set of *Partial Latin Rectangles* with $n$ rows, $r$ columns, $m$ symbols and $d$ occupied cells.

# On a 1Ă—nĂ—r chessboard

These are ordinary (2-dimensional) rook placements on rectangular chessboards. There is an extensive literature on them. The  following formula, for the placements of $d$ rooks on a $n \times r$ chessboard, is well-known and easy to derive:

$$
\# R(1,n,r|d)
\frac{(n)_d  \cdot (r)_d}{d!}
$$
which is also
$$
\binom{n}{d}  \cdot (r)_d
$$

```python
A.<x> = PolynomialRing(QQ)
B.<z, t> = PowerSeriesRing(A)

def g1(N):
    r"""
    Exponential generating series for the rook polynomials for a  1 x n x r chessboard,
    up to order N
    """
    return (z+t+x*z*t).exp(prec=N+1)
```

```python
show(g1(3))
```

Let us show all rook polynomials for $n \times r$ chessboards, when  $0 \le n, r \le N$.

```python
N = 5
f = g1(2*N)
res = {}
for (k, ell), c in f.monomial_coefficients().items():
    res[k, ell] = A(c*factorial(k)*factorial(ell))
M = matrix(A, N+1, N+1)
for (k, ell), c in res.items():
    if k <= N and ell <= N:
        M[k, ell] = c # ask for c(1) to get the number of rook placements
show(M)    
```

Total numbers of rook placements:

```python
show(M.subs({x: 1}))
```

# On a  2 Ă— n Ă— r chessboard

## Generating series

```python
A.<x> = PolynomialRing(QQ)
B.<z, t> = PowerSeriesRing(A)

def Hadamard(P, Q, N):
    r"""
    Hadamard product of two exponential series P and Q in the variables z and t,
    up to order N
    """
    return sum(P[i,j]*Q[i, j]*factorial(i)*factorial(j)*z^i*t^j 
               for (i, j) in P.exponents())

def g2(N):
    """
    Exponential generating series for the rook polynomials
    of format 2 x n x r, up to order N
    """
    return Hadamard(g1(N), g1(N), N)*(-x^2*z*t).exp(prec=N+1)

show(g2(3))    
```

```python
N = 10
f = g2(2*N)
res = {}
for (k, ell), c in f.monomial_coefficients().items():
    res[k, ell] = A(c*factorial(k)*factorial(ell))
M = matrix(A, N+1, N+1)
for (k, ell), c in res.items():
    if k <= N and ell <= N:   
        M[k, ell] = c
show(M)  
```

# On a  3 Ă— n x r chessboard

## Generating series

```python
A.<x> = PolynomialRing(QQ)
B.<z, t> = PowerSeriesRing(A)

def F(n, r, N):
    return sum(x^i/factorial(i) * (-z*t*x^2)^a/factorial(a) * falling_factorial(n+a, i) * falling_factorial(r+a, i)
               for a in [0..(N-n-r)/2] for i in [0..min(n,r) +a]) 
def g3(N):
    """
    Exponential generating series for the rook polynomials
    of format 3 x n x r
    """
    return (sum(F(n,r,N)^3 * z^n/factorial(n) * t^r/factorial(r)
              for (n, r) in IntegerListsLex(length=2, max_sum = N))
            * (2*z*t*x^3).exp(prec=N+1))
     

show(g3(3))
```

```python
N = 10
f = g3(2*N)

res = {}
for (k, ell), c in f.monomial_coefficients().items():
    res[k, ell] = A(c*factorial(k)*factorial(ell))
M = matrix(A, N+1, N+1)
for (n, r), c in res.items():
    if n <= N and r <=N:
        M[n, r] = c
show(M)  
```

## Direct Formula 

Direct formulas for the rook polynomials  and the numbers of placements of $d$ rooks on a $3 \times n \times r$ chessboard can be deduced from the generating series. 

They are implemented below.


```python
def rp(a, b, i):
    """
    Number of placements of i rooks on an a x b chessboard. 
    """
    return binomial(a, i) * falling_factorial(b, i)

def Omega(n, r, ell, d):
    return sum(mul(rp(n+x,r+x, y)/factorial(x) for (x, y) in zip((a, b, c), (i, j, k)))  
               for (i, j, k) in IntegerVectors(ell, length=3)
               for (a, b, c) in IntegerVectors(d, length=3)
    )

def R3_term(n, r, mu):
    return (Omega(n-mu.get_part(1),n-mu.get_part(1) , 
                  mu.get_part(0)-mu.get_part(1), mu.get_part(1)-mu.get_part(2))
            * (-1)^(mu.get_part(1)-mu.get_part(2))
            * 2^mu.get_part(2)/factorial(mu.get_part(2))
            * falling_factorial(n, mu.get_part(1))
            * falling_factorial(r, mu.get_part(1))
           )
    

def R3(n, r, d):
    """
    Number of placements of d rooks on a 3 x n x r chessboard
    """
    return sum(R3_term(n, r, mu) for mu in Partitions(d, max_length=3))

def RP3(n,r):
    """
    Rook Polynomial for a 3 x n x r chessboard
    """
    return sum(R3_term(n, r, mu) * x^N 
               for N in [0..3*min(n,r)] for mu in Partitions(N, max_length=3) )
    
```

# Packed rook placements

A rook placement of format $(m,n,r)$ is *packed* if 
* for all $p \in [1,m]$, there is some occupied cell in a position $(i,j,k)$ with $i=p$  
* for all $p \in [1,n]$, there is some occupied cell in a position $(i,j,k)$ with $j=p$, 
* for all $p \in [1,r]$, there is some occupied cell in a position $(i,j,k)$ with $k=p$, 

This requires, of course that all of $m$, $n$ and $r$ are $>0$, or that they are all $0$. 

Let $R^\ast_{m,n,r}{x}$ be the ordinary generating function for packed rook placements on a $m\times n \times r$ chessboard (``packed rook polynomial''). 
  Then the generating series for ordinary rook polynomials $R_{m,n,r}(x)$ and packed rook polynomials are related by the simple relation:
$$
\left(\sum_{m, n,r} R_{m, n, r}{x} \frac{u^m}{m!} \frac{z^n}{n!}\frac{t^r}{r!} \right) =
e^{z+t+u} \cdot 
\left( \sum_{m, n,r} R^\ast_{m, n, r}{x} \frac{u^m}{m!} \frac{z^n}{n!}\frac{t^r}{r!}\right).
$$

## 1 x n x r and 2 x n x r

```python
def f1(N):
    """
    Exponential generating series for the packed rook placements of format
    1 x n x r
    """
    return (x*z*t).exp(prec=N+1) - 1

def f2(N):
    """
    Exponential generating series for the packed rook placements of format
    2 x n x r
    """
    return (-z-t).exp(prec=N+1) * g2(N) - 2*f1(N) - 1
```

```python
show(f1(8))
```

```python
show(f2(4))
```

```python
N = 10
f = f2(2*N)

res = {}
for (k, ell), c in f.monomial_coefficients().items():
    res[k, ell] = A(c*factorial(k)*factorial(ell))
M = matrix(A, N+1, N+1)
for (k, ell), c in res.items():
    if k <= N and ell <=N:
        M[k, ell] = c
# other interesting options:
# c.leading_coefficient()  (get the numbers of latin rectangules on the diagonal) 
# c(1)            
# c.degree()    - get min(m*n, m*r, n*r)  
# c.valuation() 
# c(1).content() - divisibility properties
# c(1)/c.valuation()
show(M[:4, :4])    
```

## 3 x n x r

```python
def f3(N):
    """
    Exponential generating series for the packed rook placements of format
    3 x n x r
    """
    return (-z-t).exp(prec=N+1) * g3(N) - 3*f2(N) - 3*f1(N) - 1

show(f3(4))    
```

```python
N = 6
f = f3(2*N)

res = {}
for (k, ell), c in f.monomial_coefficients().items():
    res[k, ell] = A(c*factorial(k)*factorial(ell))
M = matrix(A, N+1, N+1)
for (k, ell), c in res.items():
    if k <= N and ell <=N:
        M[k, ell] = c
show(M)    
```
